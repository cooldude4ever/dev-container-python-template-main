```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def extract_data():
    print("Extraction des données sur le coût de la vie depuis Numbeo...")

    # Définir l'URL de la page
    url = "https://www.numbeo.com/quality-of-life/rankings_by_country.jsp"

    # Envoyer une requête GET pour récupérer le contenu de la page
    response = requests.get(url)

    if response.status_code == 200:
        print("Page récupérée avec succès.")
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extraire les données du tableau
        country_table = soup.find('table', {'id': 't2'}, {'class': 'stripe row-border order-column compact dataTable no-footer'})

        countries = []
        QoLIndex = []
        PPIndex = []
        SIndex = []
        HCIndex = []
        CLIndex = []
        PPtIRatio = []
        TCTIndex = []
        PIndex = []
        CIndex = []

        # Extraire les noms des pays, l'indice de qualité de vie et l'indice de pollution
        for row in country_table.find_all('tr')[1:]:  # Ignorer l'en-tête
            cols = row.find_all('td')
            if len(cols) > 1:
                country_name = cols[1].get_text(strip=True)  
                countries.append(country_name)

                index = cols[2].get_text(strip=True)
                QoLIndex.append(index)

                index = cols[3].get_text(strip=True)
                PPIndex.append(index)

                index = cols[4].get_text(strip=True)
                SIndex.append(index)

                index = cols[5].get_text(strip=True)
                HCIndex.append(index)

                index = cols[6].get_text(strip=True)
                CLIndex.append(index)

                index = cols[7].get_text(strip=True)
                PPtIRatio.append(index)

                index = cols[8].get_text(strip=True)
                TCTIndex.append(index)

                index = cols[9].get_text(strip=True)
                PIndex.append(index)

                index = cols[10].get_text(strip=True)
                CIndex.append(index)



        # Créer un DataFrame
        df = pd.DataFrame({'Pays': countries, 'Indice de qualité de vie': QoLIndex, "Indice de pouvoir d'achat": PPIndex,'Indice de sécurité': SIndex, 'Indice de santé': HCIndex, 'indice du coût de la vie': CLIndex, "Rapport prix de l'immobilier / revenu": PPtIRatio, 'Indice temps de trajet': TCTIndex, 'Indice de pollution': PIndex, 'Indice climatique': CIndex})

        # Enregistrer les données dans un fichier CSV
        filename = "countries.csv"
        df.to_csv(filename, index=False)
        print(f"Données enregistrées avec succès dans {filename}")

        # Retourner les données
        return {
            'pays': countries,
            'Indice_Qualité_de_Vie': QoLIndex,
            'Indice de pollution': PIndex,
        }
    
    else:
        raise Exception(f"Échec de la récupération des données. Code d'état : {response.status_code}")
```

```python
extract_data()
```
